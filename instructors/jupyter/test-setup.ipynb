```r
# load required libraries
library(ggplot2)
```

```r
# Create data frame with random variables
df <- data.frame(sample = LETTERS[1:10],
                 genotype = sample(c("phyB-9","phyA-211", "elf4-101", "toc1-2"), 10, replace=TRUE), 
                 light_condition = sample(c("SD","LD"), 10, replace = TRUE),
                 media = sample(c("GM-agar","GM-agar + SUC"), 10, replace=TRUE),
                 biomas = runif(10,0,0.3), 
                 starch = runif(10,3,7), 
                 sucrose = runif(10,0,2), 
                 chlorophyll =  runif(10,1,3))
```

```r
# print data frame
df
```

```r
# Create ggplot
# create vector with names for groups to use in plot
group.labs <- c("SD", "LD")

# draw plot
ggplot(df, aes(x = light_condition, y = media)) +
  geom_point(data=df,
             aes(x=genotype, y=media,size = biomas, colour = starch), alpha=.7)+
  scale_y_discrete(labels = group.labs)+
  scale_color_gradient(low = "salmon", high = "midnightblue", limits=c(0, NA))+
  coord_flip()+
  theme_bw()+
  theme(axis.ticks.length=unit(-0.1, "cm"),
        axis.text.x = element_text(margin=margin(5,5,0,5,"pt"),angle = 45, hjust = 1),
        axis.text.y = element_text(margin=margin(5,5,5,5,"pt")),
        axis.text = element_text(color = "black"),
        axis.title.x=element_blank())+
  xlab("Genotype")+
  labs(color="Starch (mg/g FW)", size="Biomas (g)")

```
